### List of Common Activation Functions with Advantages and Limitations

1. **Sigmoid (Logistic Function)**
   - **Formula:** \( \sigma(x) = \frac{1}{1 + e^{-x}} \)
   - **Advantages:**
     - Smooth gradient, preventing large jumps in output.
     - Output is in the range (0, 1), useful for probability predictions.
   - **Limitations:**
     - Causes vanishing gradients for very high/low inputs.
     - Output not zero-centered, which can slow down learning.

2. **Hyperbolic Tangent (Tanh)**
   - **Formula:** \( \tanh(x) = \frac{2}{1 + e^{-2x}} - 1 \)
   - **Advantages:**
     - Zero-centered output (-1, 1), which helps in learning.
     - Stronger gradients than sigmoid.
   - **Limitations:**
     - Still susceptible to vanishing gradients, though less so than sigmoid.

3. **ReLU (Rectified Linear Unit)**
   - **Formula:** \( \text{ReLU}(x) = \max(0, x) \)
   - **Advantages:**
     - Efficient and simple, leading to faster convergence.
     - Reduces the likelihood of vanishing gradient.
   - **Limitations:**
     - Can cause "dead neurons" where some neurons might never activate.
     - Not zero-centered.

4. **Leaky ReLU**
   - **Formula:** \( \text{Leaky ReLU}(x) = \max(0.01x, x) \)
   - **Advantages:**
     - Allows small, non-zero gradient when the input is negative, preventing dead neurons.
   - **Limitations:**
     - Introduces a small slope which might be suboptimal in some cases.

5. **Parametric ReLU (PReLU)**
   - **Formula:** \( \text{PReLU}(x) = \max(\alpha x, x) \), where \(\alpha\) is a learned parameter.
   - **Advantages:**
     - Adaptive slope for negative inputs, learned during training.
   - **Limitations:**
     - More computationally expensive due to the learning of additional parameters.

6. **ELU (Exponential Linear Unit)**
   - **Formula:** \( \text{ELU}(x) = x \) for \(x > 0\); \( \alpha(e^x - 1) \) for \(x \leq 0\)
   - **Advantages:**
     - Smooth and differentiable, avoids the dying ReLU problem.
     - Negative saturation helps with mean shift towards zero.
   - **Limitations:**
     - More computationally expensive than ReLU.

7. **Swish**
   - **Formula:** \( \text{Swish}(x) = x \cdot \sigma(x) \) where \( \sigma(x) \) is the sigmoid function.
   - **Advantages:**
     - Smooth non-monotonic function, which can improve training performance.
   - **Limitations:**
     - More computationally intensive.

8. **Softmax**
   - **Formula:** \( \text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}} \)
   - **Advantages:**
     - Outputs a probability distribution, useful for classification.
   - **Limitations:**
     - Computationally expensive, and sensitive to outliers.

### Python Code to Plot Activation Functions using Numpy, Pandas, and Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, x * alpha)

def prelu(x, alpha=0.1):
    return np.where(x > 0, x, x * alpha)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def swish(x):
    return x * sigmoid(x)

# Generate input values
x = np.linspace(-3, 3, 400)

# Create a DataFrame to hold the function values
data = {
    'x': x,
    'Sigmoid': sigmoid(x),
    'Tanh': tanh(x),
    'ReLU': relu(x),
    'Leaky ReLU': leaky_relu(x),
    'PReLU': prelu(x),
    'ELU': elu(x),
    'Swish': swish(x)
}

df = pd.DataFrame(data)

# Plot the functions
i=1
for column in df.columns[1:]:
    plt.figure(figsize=(12, 8))
    plt.plot(df['x'], df[column], label=column)
    plt.title(f'{df.columns[i]} Activation Functions')
    i+=1
    plt.xlabel('Input')
    plt.ylabel('Output')
    plt.legend()
    plt.grid(True)
    plt.show()



plt.figure(figsize=(12, 8))
for column in df.columns[1:]:
    plt.plot(df['x'], df[column], label=column)

plt.title(f'Activation Functions')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions using TensorFlow
def tf_sigmoid(x):
    return tf.keras.activations.sigmoid(x)

def tf_tanh(x):
    return tf.keras.activations.tanh(x)

def tf_relu(x):
    return tf.keras.activations.relu(x)

def tf_leaky_relu(x, alpha=0.01):
    return tf.keras.layers.LeakyReLU(alpha=alpha)(x)

def tf_prelu(x, alpha_initializer='zeros'):
    prelu_layer = tf.keras.layers.PReLU(alpha_initializer=alpha_initializer)
    return prelu_layer(x)

def tf_elu(x, alpha=1.0):
    return tf.keras.activations.elu(x, alpha=alpha)

def tf_swish(x):
    return tf.keras.activations.swish(x)

# Generate input values
x = np.linspace(-3, 3, 400)
x_tf = tf.convert_to_tensor(x)

# Compute function values
activation_functions = {
    'Sigmoid': tf_sigmoid(x_tf),
    'Tanh': tf_tanh(x_tf),
    'ReLU': tf_relu(x_tf),
    'Leaky ReLU': tf_leaky_relu(x_tf),
    'PReLU': tf_prelu(x_tf),
    'ELU': tf_elu(x_tf),
    'Swish': tf_swish(x_tf)
}

# Plot the functions
plt.figure(figsize=(12, 8))
for name, values in activation_functions.items():
    plt.plot(x, values.numpy(), label=name)

plt.title('Activation Functions (TensorFlow)')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()
```
