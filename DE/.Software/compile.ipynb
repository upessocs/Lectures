```python

cohorts=[]
attendance={}

def readandcompile(f,renamedDir="renamedInputs"):
    # print(f)
    global cohorts 
    global attendance
    # df=pd.read_excel(f,index_col='Student Id')
    df=pd.read_excel(f,index_col=None,header=None)

   
    infodf=df.iloc[0:7,0:2]
    date=infodf.iloc[5,1]
    
    duration=infodf.iloc[6,1]
    timestart=duration[0:5]
    marker=f"{date} {timestart}"
    marker=datetime.strptime(marker , '%d-%m-%Y %H:%M').strftime("%Y-%m-%d %H.%M")
    print(marker)
    # marker=date
    
    
    datadf=df.iloc[9::,[0,1,2,8]]
    datadf.columns = datadf.iloc[0]
    datadf = datadf[1:]
    datadf=datadf.set_index("Student Id")
    # print(datadf.head(2))
    datadf["Attendance"].replace({"P":1,"A":0})
    datadf["Attendance"]=datadf["Attendance"].replace({"P":1,"A":0})
    datadf.rename(columns = {"Attendance":marker}, inplace = True)
    print(datadf.head(2))
    
    
    cohortnew=list(set(datadf.iloc[:,0].tolist()))
    cohorts+=cohortnew
    cohorts=list(set(cohorts))
    
    
    for cohort in cohorts:
        print(cohort)
        if cohort in cohortnew:
            currentDf=datadf[(datadf["Cohort Name"] == cohort)]
            # print(currentDf.head(2))
            if cohort not in attendance.keys():
                attendance[cohort]=currentDf
            else:
                # print(attendance[cohort].head(2))
                # print(currentDf.iloc[2,currentDf.shape[1]-1].head(2))
                dfToJoin=currentDf.iloc[:,[currentDf.shape[1]-1]]
                # print("==============")
                # print(attendance[cohort].head(2))
                # print("==============")
                # print(dfToJoin.head(2))
                attendance[cohort] = attendance[cohort].join(dfToJoin, how='inner')
                # print("==============")
                # print(attendance[cohort].head(2))
            # attendance[cohort].set_index("Student Id")
    
    
    # Save renamed file
    subject=infodf.iloc[4,1].split(" (")[0]
    dateTime=datetime.strptime(str(infodf.iloc[5,1] +" " + infodf.iloc[6,1].split(" -")[0]) , '%d-%m-%Y %H:%M').strftime("%Y-%m-%d %H.%M")
    newRenamedFileName=subject+ " " + dateTime.replace(":",".") + ".xlsx"
    newRenamedFilePath= os.path.join(newRenamedDirPath,newRenamedFileName)
    # print(newRenamedFilePath)
    os.makedirs(newRenamedDirPath, exist_ok=True)
    df.to_excel(newRenamedFilePath,index=False,header=None)

    
    return attendance, infodf, datadf
```

```python
def calculateSummary(wdf):
    # wdf.iloc[:,3:]=wdf.iloc[:,3:].replace({"P":1,"A":0}).sort_index(axis=1)
    totalClass = wdf.iloc[:, 3:].count(axis=1)
    totalPresent = round(wdf.iloc[:, 3:].sum(axis=1),0)
    Percentage=round(totalPresent/totalClass*100,0)
    Percentage
    summary={'Total Classes':totalClass,'Total Present':totalPresent,'Percentage':Percentage}
    summarydf=pd.DataFrame.from_dict(summary)
    return summarydf
```

```python
# wb=xl.Workbook()


def cleanDir(outputDir):
    try: 
        os.system(f"rm -rf ./{outputDir}")
    except:
        pass
    try:
        os.mkdir(outputDir)
    except:
        pass

cleanDir(outputDirPath)
# os.chdir(outputDir)
for cohort in cohorts:
    outputFileName=os.path.join(outputDirPath," "+ cohort)
    print(outputFileName)
    
    
    wdf=attendance[cohort]
    
    #sort date wise
    mainData=wdf.iloc[:,3:wdf.shape[1]-1]
    sortedCols=sorted(mainData.columns)
    wdf.iloc[:,3:wdf.shape[1]-1]=mainData.reindex(sortedCols,axis=1)
    
    # wdf.iloc[0, 3:]=wdf.iloc[0, 3:].apply(pd.to_datetime)
    # wdf.iloc[:, 3:]=wdf.iloc[:, 3:].sort_values(axis=1)
    summary=calculateSummary(wdf)
    wdf=pd.concat([wdf,summary],axis=1)
    # wdf=wdf.set_index("Student Id")
    wdf=wdf.sort_values(['Name'], ascending=[True])
    wdf=wdf.reset_index()
    wdf.index += 1
    os.makedirs(outputDirPath, exist_ok=True)
    wdf.to_excel(outputFileName+".xlsx")
    wdf.to_csv(outputFileName+".csv")

    

# os.chdir("./..")
```
